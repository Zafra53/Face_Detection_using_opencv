```python
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import time
```

```python
cascPath = 'C:\\Users\\Prashant\\Desktop\\Detect\\opencv-master\\data\\haarcascades\\haarcascade_frontalface_alt.xml'
```

```python
faceCascade = cv2.CascadeClassifier(cascPath)
```

```python
def detect_faces_show(fpath):
    img = cv2.imread(fpath)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    faces = faceCascade.detectMultiScale(gray,scaleFactor=1.2,minNeighbors=2,minSize=(30,30),flags=cv2.CASCADE_SCALE_IMAGE)
    print("Found %d faces !"%len(faces))
    
    for(x,y,w,h)in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
        plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
detect_faces_show('C:\\Users\\Prashant\\Desktop\\Detect\\pic.jpg')
```
